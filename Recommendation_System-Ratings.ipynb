```python
# Import all the necessary libraries
import csv
import pandas as pd
import numpy as np 
from numpy import random
```

```python
df_1 = pd.read_csv('gdrive/My Drive/Colab Notebooks/All_YT_ratings_sum.csv')

array = df_1.to_numpy()

# Loop through each video and assign users to each video randomly and calculate the ratio of likes to dislikes for each video
for i in range(len(array)):
  title = str(array[i][0])
  likes = str(array[i][1])
  likes1 = int(array[i][1])
  dislikes = str(array[i][2])
  dislikes1 = int(array[i][2])
  userid = 'user_'+str(random.randint(0,100))

  try:
    if dislikes1 == 0:
      rating = likes1
    else:
      rating = likes1/dislikes1

  except:
    print("Something is not right here!")

  with open('gdrive/My Drive/Colab Notebooks/All_YT_ratings_ratio.csv', 'a',encoding='utf-8') as newFile:
    FileWriter = csv.writer(newFile)
    FileWriter.writerow([userid, title, likes, dislikes, rating])
```

```python
# Drop rows where either of the column values is zero
df_1 = pd.read_csv('gdrive/My Drive/Colab Notebooks/All_YT_ratings_ratio.csv')

df_2 = df_1.drop(df_1[(df_1.Likes == 0) | (df_1.Dislikes == 0)].index)

df_2.to_csv('gdrive/My Drive/Colab Notebooks/New_ratings.csv', encoding='utf-8')
```

```python
# Normalize the ratio of likes to dislikes on a scale of 0-10
from sklearn import preprocessing

ratio_videos = pd.read_csv("gdrive/My Drive/Colab Notebooks/New_ratings.csv")
scaler = preprocessing.MinMaxScaler(feature_range=(0, 10))
names = ratio_videos[["Ratio"]]
d = scaler.fit_transform(names)
scaled_df = pd.DataFrame(d, columns=["Mean"])
data_final = pd.concat([ratio_videos, scaled_df], axis = 1, join = 'inner')
scaled_df2 = data_final.loc[data_final["Mean"] > 0 ]
scaled_df2.to_csv('gdrive/My Drive/Colab Notebooks/New_ratings_scaled.csv', encoding='utf-8')
```

```python
user_ratings=pd.read_csv('gdrive/My Drive/Colab Notebooks/New_ratings_scaled.csv')
```

```python
user_ratings_pivot = user_ratings.pivot(index='userID', columns='Title', values='Mean')

print(user_ratings_pivot)

user_ratings_df=user_ratings_pivot
```

```python
# Fill rows with NaN with 0 instead
user_ratings_pivot.fillna(0, inplace=True)
user_ratings_pivot
```

```python
# Import the required libraries 
from scipy.sparse.linalg import svds
import numpy as np

# Decompose the matrix
U, sigma, Vt = svds(user_ratings_pivot)

# Convert sigma into a diagonal matrix
sigma = np.diag(sigma)
print(sigma)
```

```python
avg_ratings = user_ratings_pivot.mean(axis=1)
```

```python
# Dot product of U and sigma
U_sigma = np.dot(U, sigma)

# Dot product of result and Vt
U_sigma_Vt = np.dot(U_sigma, Vt)

# Add back on the row means contained in avg_ratings
uncentered_ratings = U_sigma_Vt + avg_ratings.values.reshape(-1, 1)

# Create DataFrame of the results
calc_pred_ratings_df = pd.DataFrame(uncentered_ratings, index=user_ratings_df.index,columns=user_ratings_df.columns)

# Print both the recalculated matrix and the original                                    
print(calc_pred_ratings_df)
```

```python
# Sort the ratings of User 8 from high to low
user_8_ratings = calc_pred_ratings_df.loc['user_8',:].sort_values(ascending=False)

print(user_8_ratings.head())
```
